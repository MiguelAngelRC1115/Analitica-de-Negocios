En este documento se desarrolla y analiza un modelo Naive Bayes para detectar lavados de activos
en una entidad financiera utilizando variables socioeconómicas y de comportamiento financiero de
los clientes. Estas variables consideradas son:

**Edad:** edad que posee el cliente cuando se hace el análisis.

**Ingresos:** ingresos declarados por el cliente en USD.

**Gastos:** gastos mensuales del cliente en USD.

**Número de tarjetas de crédito:** cantidad de tarjetas de crédito activas que posee el cliente.

**Monto transado en tarjetas:** monto total en USD que el cliente ha transado con sus tarjetas de crédito.

**Porcentaje de crecimiento patrimonial:** porcentaje de crecimiento del patrimonio del cliente.

0. Se procede con la carga de librerías de trabajo

```python
import numpy as np #Librería numérica
import pandas as pd #Librería para establecer comunicación con archivos de excel

#Librerías especificas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos de Lavado de Activos

```python
nxl= "/content/2. LavadoActivos.xlsx"
XDB= pd.read_excel(nxl, sheet_name=0)
XDB.head (100) #Muestra la base de datos cargada

#Seleccionamos las variables de trabajo
XD=XDB[['edad','ingresos_usd','gastos_usd','num_tarjetas_credito','monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio']] #Variables de trabajo
XD.head(10)
yd=XDB[['lavado_activos']]
yd.head()
```

2. Implementamos Modelo Naive Bayes

```python
mnb= GaussianNB()
mnb.fit(XD, yd) #Ajustamos las variables de entrada y salida

#Mostramos las medias de las variables
u=mnb.theta_
sigma= mnb.var_; sigma=np.sqrt(sigma)
print("'edad','ingresos_usd','gastos_usd','num_tarjetas_credito','monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio'")
print(u)
print("Las desviaciones son:")
print(sigma)
```

**ANÁLISIS DE RESULTADOS**

A partir de los resultados obtenidos, se identifica que los clientes asociados a lavado de activos presentan rasgos diferenciadores frente a aquellos que no incurren en esta actividad ilícita. En particular, estos clientes suelen exhibir patrones financieros atípicos, especialmente en variables como el crecimiento patrimonial y los valores transaccionados. Asimismo, se evidencian diferencias relevantes en el número de tarjetas de crédito y en la relación entre ingresos y gastos, lo cual constituye señales de alerta útiles para la identificación de operaciones sospechosas.

3. Se procede con la evaluación del modelo. Para la evaluación de este tipo de modelo se utiliza la matriz de confusión.
